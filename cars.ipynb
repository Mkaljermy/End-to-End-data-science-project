```python
def join_five_tables(table1, table2, table3, table4,table5, join_columns):
    # Perform inner join between the first two tables
    result = pd.merge(table1, table2, on=join_columns[0])
    result = pd.merge(result, table3, on=join_columns[1])
    result = pd.merge(result, table4, on=join_columns[2])
    result = pd.merge(result, table5, on=join_columns[3])
    
    

    return result
```

```python
# Join columns for each pair of tables
join_columns = [('IDCar','IDCar'),
                ('IDCustomer','IDCustomer'),
                ('IDDealer','IDDealer'),
                ('IDOrder','IDOrder')]
# Call the join_four_tables function
result_df = join_five_tables(data, data1, data2, data3,data4, join_columns)

# Display the result
display(result_df)
```
